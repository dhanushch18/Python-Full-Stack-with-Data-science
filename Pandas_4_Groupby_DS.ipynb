Groupby


The groupby method allows you to group rows of data together and call aggregate functions



```python
import pandas as pd
#Create dataframe
data = {'Company':['Redmi','Redmi','Apple','Apple','Samsung','Samsung'],
        'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
        'Sales':[200,120,340,124,243,350]}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df.groupby('Company')
```

```python
comp=df.groupby('Company')
```

```python
comp.mean()
```

```python
comp.std()
```

```python
comp.min()
```

```python
comp.max()
```

```python
comp.count()
```

```python
comp.describe()
```

```python
comp.describe().transpose()
```

```python
comp.describe().transpose()['Redmi']
```
